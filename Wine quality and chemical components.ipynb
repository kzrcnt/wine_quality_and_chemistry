# Química y la calidad de los vinos

## Resumen

En este proyecto se utilizó Python para analizar dos bases de datos donde se registran las cualidades químicas de varios vinos (tintos y blancos) y los efectos de éstas en su calidad. El objetivo fue encontrar correlaciones entre todas estas cualidades químicas, el contenido de alcohol y la preferencia por estos vinos. 

## Datos
Ambas bases de datos fueron extraídas del UCI Machine Learning Repository, una colección de base de datos de la Universidad de California (Irvine). Este repositorio es uno de muchos recursos disponibles en línea para científicos de datos y la comunidad de machine learning (ML).

Los datos de ambos sets corresponden a vinos de la marca portuguesa Vinho Verde. El set de tintos contiene 1,599 registros; el de blancos 4,898. Sus valores están clasificados bajo las siguientes columnas: 

- **fixed acidity**: Acidez fija, o acidez estable. Se refiere a los ácidos que permanecen en el vino después de que se caliente. Los ácidos fijos contribuyen a la estructura y complejidad del vino. Complementan el contenido alcohólico de la bebida, asegurando que ésta no se sienta ni demasiado pesada ni demasiado plana. Suele medirse en gramos de ácido tartárico por litro (g/L).
- **volatile acidity**: Acidez volátil. Mide los ácidos gaseosos en el vino, particularmente el ácido acético, característico por su aroma a vinagre. La acidez volátil suele añadir complejidad a ciertos vinos, pero también puede dar resultados negativos si se descuida. Altos niveles de acidez volátil tienden a relacionarse con “suciedad” en las uvas. Entre los factores que lo afectan se cuentan la higiene en los procesos, la edad de los barriles y ambientes oxidativos. Los vinos más dulces suelen tener niveles más altos de acidez volátil. Se mide en gramos de ácido acético por litro (g/L).
- **citric acid**: Ácido cítrico. Se añade a los vinos para aumentar su acidez, complementar sabores y prevenir “brumosidad”. Sirve como estabilizador, añadiendo el carácter cítrico del vino y proveyendo un sabor más “fresco”. Se mide en gramos por litro (g/L).
- **residual sugar**: Azúcar residual. Cantidad de azúcar que queda en el vino tras su proceso de fermentación. Es afectada por los niveles de concentración de uva y el proceso de fermentado. A mayor azúcar residual, más dulce el vino. Se mide en gramos de azúcar por litro (g/L).
- **chlorides**: Cloruros. Sus niveles son afectados por factores como el tipo de uva, la tierra en la que se cultivó, el clima y proximidad al océano. Altos niveles de cloruro en el vino le dan un sabor salado. Se mide en miligramos por litro (mg/L).
- **free sulfur dioxide**: Dióxido de azufre libre. El dióxido de azufre en el vino ayuda a prevenir su oxidación y el brote de bacterias. Demasiado dióxido de azufre libre puede inhibir los aromas del vino y contribuir a sabores amargos o metálicos. Se mide en miligramos por litro (mg/L). 
- **total sulfur dioxide**: Dióxido de azufre total. La suma de los dióxidos de azufre en el vino. Los dióxidos de azufre previenen la oxidación de los aromas del vino, elimina los sabores “viejos” y atenúan sabores rancios y mohosos. Medido en miligramos por litro (mg/L).
- **density**: Densidad. Mide la concentración de sólidos disueltos en el vino. Depende del contenido de azúcar en el vino, al igual que su contenido alcohólico y otros factores. La densidad afecta la textura, el sabor y la calidad. Medida en gramos por centímetro cúbico (gm/cm^3).
- **pH**: El pH mide los niveles de acidez en el vino. El pH afecta el sabor, estabilidad, color y otras cualidades del vino. Un pH bajo hace que el vino sepa más firme y refrescante; un pH alto le da redondez y suavidad. El pH se mide de una escala del 0 al 14. Mientras más bajo el nivel de pH, mayor es la acidez. 
- **sulphates**: Sulfitos. Compuestos químicos que ocurren natural o artificialmente en los vinos. Sirven para preservarlo, prevenir su oxidación y combatir brotes de microbios. Los sulfitos ayudan a mantener el sabor de los vinos, su color y longevidad. Medidos en miligramos por litro mg/L.
- **alcohol**: Alcohol. El contenido alcohólico se refiere al porcentaje de etanol (alcohol puro) en el vino. Las uvas de climas más cálidos producen vinos con mayores contenidos de alcohol. Los vinos ligeros y firmes tienen bajos niveles de alcohol, mientras que los más intensos tienen niveles más altos. El nivel de alcohol se expresa en porcentajes, que representan mililitros por litro. Por ejemplo, un 15% de alcohol significa que 15 mililitros de cada 100 en el vino son alcohol.
-  **quality**. Calidad. Valor que expresa la calidad del vino en una escala del 0 al 10. 

Para más detalles sobre ambas bases de datos, consulte el siguiente vínculo: https://archive.ics.uci.edu/dataset/186/wine+quality 


## Análisis

Ambas bases de datos son lo suficientemente simples y claras, así que no hizo falta hacer limpieza ni modificaciones. Las unidades de medida para cada columna (si se ocupaban) se añadieron en las visualizaciones, que son el producto final de este análisis.

Se optó por dos tipos de visualizaciones:

- Diagramas de dispersión que establecen correlaciones entre cualidades químicas, contenido de alcohol y calidad del vino. Se hizo un diagrama por cada cualidad química, exceptuando el nivel de alcohol, dando como resultado 10 diagramas.
- Diagramas de caja para ilustrar los niveles de cada cualidad química y su relación con la calidad del vino. Se hizo un diagrama por cada cualidad química, incluyendo el nivel de alcohol, resultando en 11 diagramas.

Ambos tipos de diagramas se organizaron en parrillas de 5x2.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Vinos tintos

```python
red_wine = pd.read_csv('red_wine_quality.csv')
red_wine.head()
```

```python
red_wine.describe()
```

#### Diagrama de dispersión (tintos)

```python
red_wine_fig = plt.figure(figsize=(12,17), tight_layout=True)
red_wine_fig.suptitle('Cualidades Químicas, Contenido de Alcohol y Calidad de Vinos Tintos', y=1, size='x-large', weight='demibold')

red_plot1 = plt.subplot(5, 2, 1)
sns.scatterplot(data=red_wine, x='alcohol', y='fixed acidity', hue='quality', palette='rocket')
red_plot1.set(title='Acidez Fija', xlabel='Alcohol (%)', ylabel='Acidez Fija (g/L)')

red_plot2 = plt.subplot(5, 2, 2)
sns.scatterplot(data=red_wine, x='alcohol', y='volatile acidity', hue='quality', palette='rocket')
red_plot2.set(title='Acidez Volátil', xlabel='Alcohol (%)', ylabel='Acidez Volátil (g/L)')

red_plot3 = plt.subplot(5, 2, 3)
sns.scatterplot(data=red_wine, x='alcohol', y='citric acid', hue='quality', palette='rocket')
red_plot3.set(title='Ácido Cítrico', xlabel='Alcohol (%)', ylabel='Ácido Cítrico (g/L)')

red_plot4 = plt.subplot(5, 2, 4)
sns.scatterplot(data=red_wine, x='alcohol', y='residual sugar', hue='quality', palette='rocket')
red_plot4.set(title='Azúcar Residual', xlabel='Alcohol (%)', ylabel='Azúcar Residual (g/L)')

red_plot5 = plt.subplot(5, 2, 5)
sns.scatterplot(data=red_wine, x='alcohol', y='chlorides', hue='quality', palette='rocket')
red_plot5.set(title='Cloruros', xlabel='Alcohol (%)', ylabel='Cloruros (mg/L)')

red_plot6 = plt.subplot(5, 2, 6)
sns.scatterplot(data=red_wine, x='alcohol', y='free sulfur dioxide', hue='quality', palette='rocket')
red_plot6.set(title='Dióxido de Azufre Libre', xlabel='Alcohol (%)', ylabel='Dióxido de Azufre Libre (mg/L)')

red_plot7 = plt.subplot(5, 2, 7)
sns.scatterplot(data=red_wine, x='alcohol', y='total sulfur dioxide', hue='quality', palette='rocket')
red_plot7.set(title='Dióxido de Azufre Total', xlabel='Alcohol (%)', ylabel='Dióxido de Azufre Total (mg/L)')

red_plot8 = plt.subplot(5, 2, 8)
sns.scatterplot(data=red_wine, x='alcohol', y='density', hue='quality', palette='rocket')
red_plot8.set(title='Densidad', xlabel='Alcohol (%)', ylabel='Densidad (g/cm^3)')

red_plot9 = plt.subplot(5, 2, 9)
sns.scatterplot(data=red_wine, x='alcohol', y='pH', hue='quality', palette='rocket')
red_plot9.set(title='pH', xlabel='Alcohol (%)', ylabel='pH')

red_plot10 = plt.subplot(5, 2, 10)
sns.scatterplot(data=red_wine, x='alcohol', y='sulphates', hue='quality', palette='rocket')
red_plot10.set(title='Sulfitos', xlabel='Alcohol (%)', ylabel='Sulfitos (mg/L)')

plt.savefig('red_wine_chemistry_quality_scatter.png', format='png', bbox_inches='tight')
```

Lo primero que resalta es que los vinos mejor calificados tienden a mayores porcentajes de alcohol. Los puntos más claros –tanto en este caso como en el de los blancos– se concentran en el extremo derecho de las gráficas.<br>

Parece que hay preferencia por vinos con niveles de alcohol entre 12%-14%. Esto es considerablemente más notable en los tintos, aunque esto podría atribuirse a la selección de colores para los diagramas de dispersión. <br>

Los vinos peor calificados suelen tener un nivel de alcohol de entre 10%-11%.

Se observa que se prefieren los vinos con las siguientes relaciones entre cualidades químicas y contenido de alcohol:

- **Acidez fija**: Entre ~8-12g/L, con 11-14%
- **Acidez volátil**: ~0.5-0.9g/L, 12.5-14%
- **Ácido cítrico**: ~0.3-0.6g/L, 11.5-14%
- **Azúcar residual**: ~3-6g/L, 12.5-14%
- **Cloruros**: =<0.1mg/L, 11.5-14%
- **Dióxido de azufre libre**: ~10-40mg/L, 11.5-14%
- **Dióxido de azufre total**: ~25-100mg/L, 11.5-14%
- **Densidad**: ~0.993-0.998g/cm^3, 12-13.5%
- **pH**: ~3.1-3.5, 11.5-14%
- **Sulfitos**: ~0.75-1.15mg/L, 10-14%

## Conclusiones

Parece haber una preferencia por porcentajes de alcohol que rondan 12%. En el caso de los tintos, los vinos mejor calificados tienden hacia niveles de alcohol de ~12.15%, mientras que en los blancos se prefieren niveles de ~12.50%.

En los tintos se opta por sabores ligeramente más cítricos, y en ambos parece haber un desagrado por los vinos más dulces.

El nivel de pH adecuado oscila entre 3.23 y 3.28.
